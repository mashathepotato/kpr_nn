```python
transform = transforms.Compose([
    transforms.Resize((80, 80)),
    transforms.ToTensor(),
])

dataset = data_prep.ConeDataset("data/cone_img", "data/annotations", transform=transform)
data_loader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
def plot_sample(dataset, index):
    image, keypoints = dataset[index]
    # print(image, keypoints)
    image = image.permute(1, 2, 0).numpy()
    keypoints = keypoints.numpy()
    
    plt.imshow(image)
    plt.scatter(keypoints[:, 0], keypoints[:, 1], c='red', marker='o')
    plt.show()

plot_sample(dataset, 0)
```
